# ファイルのインストール

```python
import pandas as pd
from rdkit.Chem import PandasTools
```

```python
file_name = "/content/DPPH_kitamura1.csv"

df = pd.read_csv(file_name)

PandasTools.AddMoleculeColumnToFrame(df, "SMILES")

df
```

```python
from mordred import Calculator, descriptors
```

```python
calc = Calculator(descriptors, ignore_3D=True)
```

```python
df_mordred = calc.pandas(df["ROMol"])
```

```python
df_mordred.head()
```

```python
df_comb = df.merge(df_mordred, left_index=True, right_index=True)
df_comb
```

```python
df_comb.to_csv('output.csv', index=False)
```
